```python
# Example of for loop
fruits = ['apple', 'banana', 'orange', 'grape']

print("Printing fruits using for loop:")
for fruit in fruits:
    print(fruit)

# Iterating over a range of numbers
print("\nPrinting numbers from 1 to 5 using for loop:")
for number in range(1, 6):
    print(number)

# Example of while loop
counter = 0

print("\nPrinting numbers using while loop:")
while counter < 5:
    print(counter)
    counter += 1

# Using break statement to exit a loop
print("\nUsing break statement to exit a loop:")
for i in range(10):
    if i == 5:
        break
    print(i)

# Using continue statement to skip an iteration
print("\nUsing continue statement to skip an iteration:")
for i in range(5):
    if i == 2:
        continue
    print(i)

# Nested loops
print("\nNested loops:")
for i in range(3):
    for j in range(2):
        print(f"({i}, {j})")

# Looping through dictionary items
student_info = {'name': 'Alice', 'age': 20, 'grade': 'A'}
print("\nLooping through dictionary items:")
for key, value in student_info.items():
    print(f"{key}: {value}")
```
