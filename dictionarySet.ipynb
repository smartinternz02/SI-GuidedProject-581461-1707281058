```python
# Dictionary example
student_info = {
    'name': 'Alice',
    'age': 20,
    'grades': {'math': 90, 'history': 85, 'english': 92}
}

# Accessing values in a dictionary
print("Student name:", student_info['name'])
print("Student age:", student_info['age'])
print("Math grade:", student_info['grades']['math'])

# Modifying values
student_info['age'] = 21
student_info['grades']['history'] = 88
print("Updated student info:", student_info)

# Adding new key-value pairs
student_info['gender'] = 'female'
print("Student info with gender:", student_info)

# Removing a key-value pair
removed_grade = student_info['grades'].pop('english')
print("Removed English grade:", removed_grade)
print("Updated student info after removing English grade:", student_info)

# Iterating through keys and values
print("Iterating through keys:")
for key in student_info:
    print(key)

print("\nIterating through values:")
for value in student_info.values():
    print(value)

# Set example
unique_numbers = {1, 2, 3, 4, 5}

# Adding elements to a set
unique_numbers.add(6)
unique_numbers.add(3)  # Adding a duplicate does not change the set
print("Updated set of unique numbers:", unique_numbers)

# Removing elements from a set
unique_numbers.remove(2)
print("Set after removing 2:", unique_numbers)

# Set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union of sets
union_result = set_a.union(set_b)
print("Union of sets:", union_result)

# Intersection of sets
intersection_result = set_a.intersection(set_b)
print("Intersection of sets:", intersection_result)

# Difference between sets
difference_result = set_a.difference(set_b)
print("Difference between sets:", difference_result)
```
