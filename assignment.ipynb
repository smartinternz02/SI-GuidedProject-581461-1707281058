```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Iris dataset from Seaborn

```python
iris = sns.load_dataset('iris')
```

```python
print("Iris Dataset:")
print(iris.head())
```

```python
sepal_length = np.array(iris['sepal_length'])
sepal_width = np.array(iris['sepal_width'])
petal_length = np.array(iris['petal_length'])
petal_width = np.array(iris['petal_width'])
```

```python
mean_sepal_length = np.mean(sepal_length)
max_petal_width = np.max(petal_width)
min_petal_length = np.min(petal_length)
```

```python

print("\nNumPy Operations:")
print(f"Mean Sepal Length: {mean_sepal_length}")
print(f"Max Petal Width: {max_petal_width}")
print(f"Min Petal Length: {min_petal_length}")
```

```python
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Box Plot of Sepal Lengths by Species")
plt.show()
```
